## Nodes (Vertices):

Each node represents a bank account or a customer.

## Edges:

An edge connects two nodes if there is a transaction between the two accounts.

Example edges:

Alisha → Bobby (₹500)


### Interpretation:

The graph shows how money flows between accounts.

Accounts with many edges handle more transactions.

Useful for analyzing transaction patterns or detecting suspicious activity.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph (transactions have direction)
G = nx.DiGraph()

# Add nodes (bank accounts)
accounts = ["Alisha", "Bobby", "Chartan", "Darshan", "Esha"]
G.add_nodes_from(accounts)

# Add edges with transaction amounts as weights
transactions = [
    ("Alisha", "Bobby", 500),
    ("Bobby", "Chartan", 1000),
    ("Chartan", "Alisha", 300),
    ("Darshan", "Esha", 700),
    ("Esha", "Bobby", 200)
]

for sender, receiver, amount in transactions:
    G.add_edge(sender, receiver, weight=amount)

# Draw the graph
plt.figure(figsize=(8,8))
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color="palegreen", node_size=2000, edgecolors="black")
# Draw edges with arrows
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, width=2)
# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=10)
# Draw edge labels (transaction amounts)
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="blue")

plt.title("Bank Transaction Network", fontsize=14, fontweight="bold")
plt.axis("off")
plt.show()
```
