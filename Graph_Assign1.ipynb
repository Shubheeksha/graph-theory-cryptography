### Real-World Situation:
We want to represent artist collaborations in the music industry as a graph.

### Nodes (Vertices):

Each node represents an artist.<br>
Example: Ed Sheeran, Justin Bieber, Eminem.



### Edges (Connections):

An edge connects two artists if they have collaborated on at least one song.<br>
Example edges:<br>
Ed Sheeran - Eminem (River)


## Graphical Representation:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes (artists)
artists = ["Ed Sheeran", "Justin Bieber", "Eminem", "Ariana Grande", "Taylor Swift","DJ Snake"]
G.add_nodes_from(artists)

# Add edges with song collaborations
collaborations = [
    ("Ed Sheeran", "Justin Bieber", "I Dont Care"),
    ("Ed Sheeran", "Eminem", "River"),
    ("Ariana Grande", "Justin Bieber", "Stuck With U"),
    ("Taylor Swift", "Ed Sheeran", "Everything Has Changed"),
    ("Justin Bieber", "DJ Snake", "Let Me Love You")
]

for artist1, artist2, song in collaborations:
    G.add_edge(artist1, artist2, song=song)

# Draw the graph
plt.figure(figsize=(12,12))
pos = nx.spring_layout(G, seed=42)  # Positions nodes nicely

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_color="skyblue", node_size=4500, edgecolors="black")
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=7, font_weight="bold")

# Draw edge labels (song names)
edge_labels = nx.get_edge_attributes(G, "song")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red", font_size=9)

plt.title("Artist Collaboration Graph", fontsize=14, fontweight="bold")
plt.axis("off")
plt.show()
```

## The Graph Shows:

Nodes: Artists

Edges: Collaborative songs connecting two artists

Edge labels: Name of the song they collaborated on

### This graph helps identify:

Popular collaborators (nodes with many edges)

How artists are connected in the industry

Bridges between different artist groups
